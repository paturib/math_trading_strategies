```python
import yfinance as yf
import matplotlib.pyplot as plt
```

## Descending Triangle

```python
def identify_descending_triangle(data):
    highs = data["High"]
    lows = data["Low"]

    # Find support and resistance levels
    resistance_line = highs[highs.idxmax():]
    support_line = lows[highs.idxmax():]

    # Plot the descending triangle pattern
    plt.plot(resistance_line, color="red", linestyle="-", linewidth=2, label="Resistance")
    plt.plot(support_line, color="green", linestyle="-", linewidth=2, label="Support")
    plt.legend() 
```

## Bull Flag

```python
def identify_bull_flag(data):
    highs = data["High"]
    lows = data["Low"]

    # Find the flagpole and consolidation range
    flagpole = highs[highs.idxmax():]
    consolidation = lows[highs.idxmax():]

    # Plot the bull flag pattern
    plt.plot(flagpole, color="red", linestyle="-", linewidth=2, label="Flagpole")
    plt.plot(consolidation, color="green", linestyle="-", linewidth=2, label="Consolidation")
    plt.legend()
```

## Inverse Cup and Handle

```python
def identify_inverse_cup_and_handle(data):
    highs = data["High"]
    lows = data["Low"]

    # Find the inverse cup and handle pattern
    cup = highs[highs.idxmax():]
    handle = lows[highs.idxmax():]

    # Plot the inverse cup and handle pattern
    plt.plot(cup, color="red", linestyle="-", linewidth=2, label="Cup")
    plt.plot(handle, color="green", linestyle="-", linewidth=2, label="Handle")
    plt.legend()
```

## Inverse Head and Shoulder

```python
def identify_inverse_head_and_shoulder(data):
    highs = data["High"]
    lows = data["Low"]

    # Find the inverse head and shoulder pattern
    head = highs[highs.idxmax():]
    shoulders = lows[highs.idxmax():]

    # Plot the inverse head and shoulder pattern
    plt.plot(head, color="red", linestyle="-", linewidth=2, label="Head")
    plt.plot(shoulders, color="green", linestyle="-", linewidth=2, label="Shoulders")
    plt.legend()
```

## Rounding Bottom

```python
def identify_rounding_bottom(data):
    lows = data["Low"]

    # Find the rounding bottom pattern
    rounding_bottom = lows[lows.idxmin():]

    # Plot the rounding bottom pattern
    plt.plot(rounding_bottom, color="blue", linestyle="-", linewidth=2, label="Rounding Bottom")
    plt.legend()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the stock ticker symbol
ticker = "MSFT"

# Fetch historical data using yfinance
data = yf.download(ticker, period="10y", interval="1d")

# Function to identify and plot the descending triangle pattern
def identify_descending_triangle(data):
    highs = data["High"]
    lows = data["Low"]

    # Find the support and resistance levels
    resistance_line = highs[highs.idxmax():]
    support_line = lows[highs.idxmax():]

    # Plot the descending triangle pattern
    plt.plot(data["Close"], color="blue", linestyle="-", linewidth=2, label="Closing Price")
    plt.plot(resistance_line, color="red", linestyle="-", linewidth=2, label="Resistance")
    plt.plot(support_line, color="green", linestyle="-", linewidth=2, label="Support")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.title(f"Stock: {ticker}")
    plt.legend()
    plt.show()

    # Identify the start of the descending triangle pattern
    start_date = resistance_line.index[0]
    print(f"The descending triangle pattern started on: {start_date}")

# Call the function to identify and plot the descending triangle pattern
identify_descending_triangle(data)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the stock ticker symbol
ticker = "AAPL"

# Fetch historical data using yfinance
data = yf.download(ticker, period="10y", interval="1d")

# Function to identify and plot the bull flag pattern
def identify_bull_flag(data):
    highs = data["High"]
    lows = data["Low"]

    # Find the flagpole and consolidation range
    flagpole = highs[highs.idxmax():]
    consolidation = lows[highs.idxmax():]

    # Plot the bull flag pattern
    plt.plot(data["Close"], color="blue", linestyle="-", linewidth=2, label="Closing Price")
    plt.plot(flagpole, color="red", linestyle="-", linewidth=2, label="Flagpole")
    plt.plot(consolidation, color="green", linestyle="-", linewidth=2, label="Consolidation")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.title(f"Stock: {ticker}")
    plt.legend()
    plt.show()

    # Identify the start of the bull flag pattern
    start_date = flagpole.index[0]
    print(f"The bull flag pattern started on: {start_date}")

# Call the function to identify and plot the bull flag pattern
identify_bull_flag(data)
```
